```python
# We'll start off by dropping the Cabin feature since not a lot of useful information can be extracted from it
# as we have seen from insights.
dtrain = dtrain.drop(['Cabin'], axis = 1)
dtest = dtest.drop(['Cabin'], axis = 1)
```

```python
# We can also drop the Ticket feature since it's unlikely to yield any useful information
dtrain = dtrain.drop(['Ticket'], axis = 1)
dtest = dtest.drop(['Ticket'], axis = 1)
```

```python
# Now we need to fill in the missing values in the Embarked feature

print("Number of people embarking in Southampton (S):")
southampton = dtrain[dtrain["Embarked"] == "S"].shape[0]
print(southampton)

print("Number of people embarking in Cherbourg (C):")
cherbourg = dtrain[dtrain["Embarked"] == "C"].shape[0]
print(cherbourg)

print("Number of people embarking in Queenstown (Q):")
queenstown = dtrain[dtrain["Embarked"] == "Q"].shape[0]
print(queenstown)
```

```python
# It's clear that the majority of people embarked in Southampton (S). 
# So Let's go ahead and fill in the missing values with S.
# This will not affect our model at all because they are very less in nunmber.
# Replacing the missing values in the Embarked feature with S
dtrain = dtrain.fillna({"Embarked": "S"})
dtrain.head()
```

```python
# Now let's deal with the Age Feature. This is not so easy.
# Next we'll fill in the missing values in the Age feature. 
# Since a higher percentage of values are missing,
# it would be illogical to fill all of them with the same value (as we did with Embarked).
# Instead, let's try to find a way to predict the missing ages.
```

```python
# create a combined group of both datasets
combine = [dtrain, dtest]

# extract a title for each Name in the train and test datasets
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.')

pd.crosstab(dtrain['Title'], dtrain['Sex'])

```

```python
# To Generalize,let's replace various titles with more common names

for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Capt', 'Col',
    'Don', 'Dr', 'Major', 'Rev', 'Jonkheer', 'Dona'], 'Rare')
    
    dataset['Title'] = dataset['Title'].replace(['Countess', 'Lady', 'Sir'], 'Royal')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

dtrain[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
# Now let's map each of the title groups to a numerical value
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Royal": 5, "Rare": 6}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

dtrain.head()
```

```python
# Now, we'll try to predict the missing Age values from the most common age for their Title.
```

```python
# Now let's fill missing age with mode age group for each title

mr_age = dtrain[dtrain["Title"] == 1]["AgeGroup"].mode() #Young Adult
miss_age = dtrain[dtrain["Title"] == 2]["AgeGroup"].mode() #Student
mrs_age = dtrain[dtrain["Title"] == 3]["AgeGroup"].mode() #Adult
master_age = dtrain[dtrain["Title"] == 4]["AgeGroup"].mode() #Baby
royal_age = dtrain[dtrain["Title"] == 5]["AgeGroup"].mode() #Adult
rare_age = dtrain[dtrain["Title"] == 6]["AgeGroup"].mode() #Adult

def age_title_mapping(title):
    #{ 1: "Young Adult", 2: "Student", 3: "Adult", 4: "Baby", 5: "Adult", 6: "Adult"}
    if title==1:
        return "Young Adult"
    elif title==2:
        return "Student"
    elif title==3:
        return "Adult"
    elif title==4:
        return "Baby"
    elif title==5 or title==6:
        return "Adult"
        
    
    
    
    
for x in range(len(dtrain["AgeGroup"])):
    if dtrain["AgeGroup"][x] == "Unknown":
        dtrain["AgeGroup"][x] = age_title_mapping(dtrain["Title"][x])
        
for x in range(len(dtest["AgeGroup"])):
    if dtest["AgeGroup"][x] == "Unknown":
        dtest["AgeGroup"][x] = age_title_mapping(dtest["Title"][x])
        
dtrain.head()
```

```python
# We've filled in the missing values at least somewhat accurately but it is way better than just replacing all 
# the missing values by a single value such as median, etc.
```

```python
# It's time to map each age group to a numerical value.
# Map each Age value to a numerical value

age_mapping = {'Baby': 1, 'Child': 2, 'Teenager': 3, 'Student': 4, 'Young Adult': 5, 'Adult': 6, 'Senior': 7}
dtrain['AgeGroup'] = dtrain['AgeGroup'].map(age_mapping)
dtest['AgeGroup'] = dtest['AgeGroup'].map(age_mapping)

dtrain.head()

```

```python
# Dropping the Age feature 
dtrain = dtrain.drop(['Age'], axis = 1)
dtest = dtest.drop(['Age'], axis = 1)
```

```python
# We can drop the name feature since it contains no more useful information.
dtrain = dtrain.drop(['Name'], axis = 1)
dtest = dtest.drop(['Name'], axis = 1)
```

```python
dtrain.head()
```

```python
# Let's map each Sex value to a numerical value
sex_mapping = {"male": 0, "female": 1}
dtrain['Sex'] = dtrain['Sex'].map(sex_mapping)
dtest['Sex'] = dtest['Sex'].map(sex_mapping)

dtrain.head()
```

```python
# Let's map each Embarked value to a numerical value
embarked_mapping = {"S": 1, "C": 2, "Q": 3}
dtrain['Embarked'] = dtrain['Embarked'].map(embarked_mapping)
dtest['Embarked'] = dtest['Embarked'].map(embarked_mapping)

dtrain.head()
```

```python
# During Insights, it was observed that people who are alone survived albeit it is sad that's the harsh truth
# So it is worth creating a new feature in it's case, that is, Alone
# Albeit, it is sad but who would travel on a cruise alone...that's just sad but hey they survived
# So as they say, every Cloud has a Silver Lining
```

```python
dtrain['Alone']=((dtrain.SibSp + dtrain.Parch)==0).astype(int)
dtrain.head()
```

```python
# Let's repeat the process for Testing data
dtest['Alone']=((dtest.SibSp + dtest.Parch)==0).astype(int)
dtest.head()
```

```python
#It's time separate the fare values into some 
# logical groups as well as filling in the single 
# missing value in the test dataset.

# Fill in missing Fare value in test set based on mean fare for that Pclass 
for x in range(len(dtest["Fare"])):
    if pd.isnull(dtest["Fare"][x]):
        pclass = dtest["Pclass"][x] #Pclass = 3
        dtest["Fare"][x] = round(dtrain[dtrain["Pclass"] == pclass]["Fare"].mean(), 4)
        
```

```python
# Let's map Fare values into groups of numerical values
dtrain['FareBand'] = pd.qcut(dtrain['Fare'], 4, labels = [1, 2, 3, 4])
dtest['FareBand'] = pd.qcut(dtest['Fare'], 4, labels = [1, 2, 3, 4])
```

```python
# I am going to be using different models to predict the results of the dataset and compare their accuracies 
```

```python
# Prepping the data to be fed to the model
predictors = dtrain.drop(['Survived', 'PassengerId'], axis=1)
target = dtrain["Survived"]
```

```python
# To measure accuracy of our prediction using various models
from sklearn.metrics import accuracy_score
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB

gaussian = GaussianNB()
gaussian.fit(predictors,target)
pred=gaussian.predict(predictors)
acc_gaussian = round(accuracy_score(pred,target) * 100, 2)
print(acc_gaussian)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(predictors, target)
pred = logreg.predict(predictors)
acc_logreg = round(accuracy_score(pred, target) * 100, 2)
print(acc_logreg)
```

```python
# Support Vector Machines
from sklearn.svm import SVC

svc = SVC()
svc.fit(predictors, target)
pred = svc.predict(predictors)
acc_svc = round(accuracy_score(pred, target) * 100, 2)
print(acc_svc)
```

```python
# Linear SVC
from sklearn.svm import LinearSVC

linear_svc = LinearSVC()
linear_svc.fit(predictors, target)
pred = linear_svc.predict(predictors)
acc_linear_svc = round(accuracy_score(pred, target) * 100, 2)
print(acc_linear_svc)
```

```python
# Perceptron
from sklearn.linear_model import Perceptron

perceptron = Perceptron()
perceptron.fit(predictors, target)
pred = perceptron.predict(predictors)
acc_perceptron = round(accuracy_score(pred, target) * 100, 2)
print(acc_perceptron)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier

decisiontree = DecisionTreeClassifier()
decisiontree.fit(predictors, target)
pred = decisiontree.predict(predictors)
acc_decisiontree = (accuracy_score(pred, target) * 100)
print(acc_decisiontree)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(predictors,target)
pred = randomforest.predict(predictors)
acc_randomforest = (accuracy_score(pred, target) * 100)
print(acc_randomforest)
```

```python
# KNN or k-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(predictors,target)
pred = knn.predict(predictors)
acc_knn = round(accuracy_score(pred, target) * 100, 2)
print(acc_knn)
```

```python
# Stochastic Gradient Descent
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()
sgd.fit(predictors,target)
pred = sgd.predict(predictors)
acc_sgd = round(accuracy_score(pred, target) * 100, 2)
print(acc_sgd)
```

```python
# Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

gbk = GradientBoostingClassifier()
gbk.fit(predictors, target)
pred = gbk.predict(predictors)
acc_gbk = round(accuracy_score(pred, target) * 100, 2)
print(acc_gbk)
```

```python
# Let us compare the models
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 'Linear SVC', 
              'Decision Tree', 'Stochastic Gradient Descent', 'Gradient Boosting Classifier'],
    'Accuracy': [acc_svc, acc_knn, acc_logreg, 
              acc_randomforest, acc_gaussian, acc_perceptron,acc_linear_svc, acc_decisiontree,
              acc_sgd, acc_gbk]})

# Print them in descending order
models.sort_values(by='Accuracy', ascending=False)
```
